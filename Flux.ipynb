```python
!pip install gradio_client

import base64
from gradio_client import Client

# Install gradio_client if not already installed

# Initialize the client with the correct endpoint
client = Client("black-forest-labs/FLUX.1-schnell")

# Generate the image with the desired prompt and parameters
result = client.predict(
    prompt="A vase inspired by the dancing house in Prague, Czech Republic designed by Vlado Milunic",
    seed=0,
    randomize_seed=True,
    width=1024,
    height=1024,
    api_name="/infer"
)

# Get the path to the generated image
result_path = result[0]  # The path to the image

# Read the image and encode it to base64
with open(result_path, "rb") as image_file:
    base64_image = base64.b64encode(image_file.read()).decode('utf-8')

# Print the base64 image string
print(f"data:image/webp;base64,{base64_image}")

# If you want to display the image in a Jupyter Notebook
from IPython.display import display, HTML
display(HTML(f"<img src='data:image/webp;base64,{base64_image}'/>"))
```
